```python
import numpy as np
import pandas as pd
from unidecode import unidecode
import re
import datetime

orders = pd.read_csv(r"Z:\Downloads\archive (13)\olist_orders_dataset.csv")

orders['order_purchase_timestamp'] = orders['order_purchase_timestamp'].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d %H:%M:%S"))
orders['order_approved_at'] = orders['order_approved_at'].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d %H:%M:%S") if not pd.isna(x) else x)
orders['order_delivered_carrier_date'] = orders['order_delivered_carrier_date'].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d %H:%M:%S") if not pd.isna(x) else x)
orders['order_delivered_customer_date'] = orders['order_delivered_customer_date'].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d %H:%M:%S") if not pd.isna(x) else x)
orders['order_estimated_delivery_date'] = orders['order_estimated_delivery_date'].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d %H:%M:%S") if not pd.isna(x) else x)

orders = orders[(
    ((orders['order_approved_at'] - orders['order_purchase_timestamp']).dt.total_seconds() / 86400 >= 0) &
    ((orders['order_delivered_carrier_date'] - orders['order_approved_at']).dt.total_seconds() / 86400 >= 0))]

orders.reset_index(drop=True, inplace=True)

to_remove = orders[(orders['order_status']=='canceled') & (orders['order_delivered_customer_date'].notna())].index

orders.drop(to_remove, inplace=True)

orders.reset_index(drop=True, inplace=True)

orders['order_status'] = orders['order_status'].str.replace('canceled', 'cancelled')

payments = pd.read_csv(r"Z:\Downloads\archive (13)\olist_order_payments_dataset.csv")

payments.drop(columns = ['payment_sequential', 'payment_installments'], inplace=True)

payments = payments.groupby(by=['order_id', 'payment_type']).sum('payment_value').reset_index()


merged_orders_payments = pd.merge(orders, payments, how='inner', on='order_id')

order_items = pd.read_csv(r"Z:\Downloads\archive (13)\olist_order_items_dataset.csv")

merged_orders_payments_orderItems = pd.merge(merged_orders_payments, order_items, how='inner', on='order_id')

merged_orders_payments_orderItems['shipping_limit_date'] = \
        merged_orders_payments_orderItems['shipping_limit_date'].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d %H:%M:%S"))

merged_orders_payments_orderItems = merged_orders_payments_orderItems[(merged_orders_payments_orderItems['shipping_limit_date'] > merged_orders_payments_orderItems['order_purchase_timestamp']) & \
(merged_orders_payments_orderItems['shipping_limit_date'] > merged_orders_payments_orderItems['order_approved_at']) & \
(merged_orders_payments_orderItems['shipping_limit_date'] < merged_orders_payments_orderItems['order_delivered_customer_date'])]

merged_orders_payments_orderItems.drop(columns=['order_item_id'], inplace=True)

merged_orders_payments_orderItems = merged_orders_payments_orderItems.groupby(['order_id', 'customer_id', 'order_status', 'order_purchase_timestamp',
       'order_approved_at', 'order_delivered_carrier_date',
       'order_delivered_customer_date', 'order_estimated_delivery_date',
       'payment_type', 'payment_value', 'product_id', 'seller_id',
       'shipping_limit_date']).agg(price=("price", "sum"),
                                    freight_value=("freight_value", "sum"),
                                      quantity=("product_id", "count")).reset_index()

to_remove = merged_orders_payments_orderItems.loc[np.where(round((merged_orders_payments_orderItems['price'] + merged_orders_payments_orderItems['freight_value']), 2) > merged_orders_payments_orderItems['payment_value'])].index

merged_orders_payments_orderItems.drop(to_remove, inplace=True)

merged_orders_payments_orderItems.reset_index(drop=True, inplace=True)

reviews = pd.read_csv(r"Z:\Downloads\archive (13)\olist_order_reviews_dataset.csv")

to_remove = reviews[~reviews['order_id'].isin(merged_orders_payments_orderItems['order_id'])].index

reviews.drop(to_remove, inplace = True)

reviews.reset_index(drop=True, inplace=True)
```
